```python
df2 = df.drop(['Unnamed: 0','date','day_of_week','functional_classification_name'],axis=1)

df3=df2.groupby(['month_of_data']).sum()
df3
```

```python
df3['total'] = df3.sum(axis=1)
df3
```

```python
axx = df3['total'].plot(kind='line',figsize=(15, 10), fontsize=12)
plt.xticks(df3.index)
axx.set_title("Total Traffic Volume per Month for Urban Roads", fontsize=20)
axx.set_xlabel("Month",fontsize=16)
axx.set_ylabel("Traffic Volume in Billions", fontsize=16)
axx.set_xticklabels(['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec'])
plt.show()
```
